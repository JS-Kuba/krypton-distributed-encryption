```python
from Crypto.Cipher import ARC4
import os

# Read the plaintext from the file
with open('../data/crime-and-punishment.txt', 'r') as f:
    plaintext = f.read()

# Replicate the file in a loop
with open('../data/crime-and-punishment-long.txt', 'w') as f:
    for x in range(100):
        f.write(plaintext)

with open('../data/crime-and-punishment-long.txt', 'r') as f:
    longtext = f.read()
```

```python
# Encrypt the text using RC4
key = b'Sixteen byte key' 
cipher = ARC4.new(key)

ciphertext = cipher.encrypt(longtext)

# Write the encrypted text to a file
with open('../data/crime-and-punishment-encrypted.txt', 'wb') as f:
    f.write(ciphertext)

# Read the encrypted text from the file
with open('../data/crime-and-punishment-encrypted.txt', 'rb') as f:
    encrypted_text = f.read()

# Create a new cipher instance with the same key and decrypt the text
cipher = ARC4.new(key)
decrypted_text = cipher.decrypt(encrypted_text)

# Print the decrypted text
#print(decrypted_text.decode('utf-8'))
```

```python
def get_bytes(text, encoding="utf-8"):
    return len(text.encode(encoding))

def split_file_to_list(file_path, block_size=500 * 1024):
    if not os.path.isfile(file_path):
        print("Error: File not found.")
        return

    blocks = []
    with open(file_path, "r", encoding="utf-8") as input_file:
        block_index = 0
        block_data = ""
        total_bytes = 0

        for line in input_file:
            line_bytes = get_bytes(line)
            total_bytes += line_bytes

            if total_bytes >= block_size:
                block_dict = {"index": block_index, "data": block_data}
                blocks.append(block_dict)
                print(f"Created block {block_index}")
                block_index += 1

                block_data = line
                total_bytes = line_bytes
            else:
                block_data += line

        if block_data:
            block_dict = {"index": block_index, "data": block_data}
            blocks.append(block_dict)
            print(f"Created block {block_index}")

    return blocks

input_file_path = '../data/crime-and-punishment-long.txt'
blocks_list = split_file_to_list(input_file_path)
```
